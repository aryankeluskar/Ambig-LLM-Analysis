```python
# go through sample.json and ask questions to gpt
import json
sample = json.load(open("sample.json"))
for i in sample:
    answers = []
    print(i["nq_question"])
    for _ in range(3):
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": i["nq_question"],
                }
            ],
            model="gpt-4o-mini", temperature=0.2
        )
        answers.append(chat_completion.choices[0].message.content)
        # print("-"*20)
    print("-"*20)
    
    # show the difference between the answers, characterwise
    for a in answers:
        print(a)

    print("="*20)
    

# default gives a detailed answer for all questions
```

```python
# go through sample.json and ask questions to gpt
import json
sample = json.load(open("sample.json"))
for i in sample:
    answers = []
    print(i["nq_question"])
    print("-"*20)
    for _ in range(3):
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": f"Add extra information to the following question. Your aim is to disambiguate what it is asking: {i['nq_question']}",
                }
            ],
            model="gpt-4o-mini", temperature=0.2
        )

        print(chat_completion.choices[0].message.content)


        final_answer = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": chat_completion.choices[0].message.content,
                }
            ],
            model="gpt-4o-mini", temperature=0.2
        )

        answers.append(final_answer.choices[0].message.content)
        # print("-"*20)
    print("-"*20)
    
    # show the difference between the answers, characterwise
    for a in answers:
        print(a)

    print("="*20)
```
