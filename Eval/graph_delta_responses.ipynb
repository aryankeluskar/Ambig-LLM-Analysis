```python
import json
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the first dataset
with open('/Users/aryank/Developer/dmml-research/Experiments/eval_o_ques_what_openai_similarity_out.json') as f:
    data1 = json.load(f)

disambig_answer_distance1 = []
ambig_answer_distance1 = []

for i in data1:
    disambig_answer_distance1.append(i['disambig_answer_distance'])
    ambig_answer_distance1.append(i['ambig_answer_distance'])

disambig_answer_distance1 = np.array(disambig_answer_distance1)
ambig_answer_distance1 = np.array(ambig_answer_distance1)

diff1 = disambig_answer_distance1 - ambig_answer_distance1
diff1 = diff1[(diff1 > 0.01) | (diff1 < -0.01)]

# Load the second dataset
with open('/Users/aryank/Developer/dmml-research/Experiments/eval_o_add_context_openai_similarity_out.json') as f:
    data2 = json.load(f)

disambig_answer_distance2 = []
ambig_answer_distance2 = []

for i in data2:
    disambig_answer_distance2.append(i['disambig_answer_distance'])
    ambig_answer_distance2.append(i['ambig_answer_distance'])

disambig_answer_distance2 = np.array(disambig_answer_distance2)
ambig_answer_distance2 = np.array(ambig_answer_distance2)

diff2 = disambig_answer_distance2 - ambig_answer_distance2
diff2 = diff2[(diff2 > 0.01) | (diff2 < -0.01)]

# Load the third dataset
with open('/Users/aryank/Developer/dmml-research/off-track exps/disambig_filter_with_our_sample.json') as f:
    data3 = json.load(f)

disambig_answer_distance3 = []
ambig_answer_distance3 = []

for i in data3:
    disambig_answer_distance3.append(i['disambig_answer_distance'])
    ambig_answer_distance3.append(i['ambig_answer_distance'])

disambig_answer_distance3 = np.array(disambig_answer_distance3)
ambig_answer_distance3 = np.array(ambig_answer_distance3)

diff3 = disambig_answer_distance3 - ambig_answer_distance3
diff3 = diff3[(diff3 > 0.01) | (diff3 < -0.01)]

# Plot the graphs on top of each other
plt.figure(figsize=(10, 6))

# scale y-axis to 2.0
# plt.ylim(0, 2.0)

plt.ylabel("Probability Density of Cosine Similarity")

# sns.kdeplot(diff1, fill=True, alpha=0.05, label='Rephrasing the question', color='blue')
sns.kdeplot(diff2, fill=True, alpha=0.2, label='Using add-context prompt', color='red')
sns.kdeplot(diff3, fill=True, alpha=0.2, label='Using human-provided disambig. question', color='green')
plt.xlabel('Semantic Distance between LLM Response and Ground Truth Answer')

# put a dotted vertical line at 0
plt.axvline(x=0, linestyle='--', color='black', alpha=0.3)

plt.legend()
plt.show()
```
