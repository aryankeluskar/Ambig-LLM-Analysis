```python
system_prompt = """
You are an expert in answering questions. Here are some examples of ambiguous questions rephrased to be more specific:

Ambiguous: Who has the authority to impeach the president?
Disambiguated: What entity has the authority to impeach the president?

Ambiguous: When was the company founded?
Disambiguated: What year was the company founded?

Now, please answer the following ambiguous questions after you rephrase them to be more specific:
"""

# This approach does not work. It is supposed to interally disambiguate the questions, but it does not follow the instructions.
# It's far better to use two different Completitions API Calls, one for disambiguation and one for answering the question.
```

```python
import os
from openai import OpenAI
import dotenv
dotenv.load_dotenv()

client = OpenAI(
  api_key=os.environ.get("OPENAI_API_KEY"),
)
```

```python
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": system_prompt,
        },
        {
            "role": "user",
            "content": "",
        }
    ],
    model="gpt-4o-2024-08-06",
)

str(chat_completion.choices[0].message.content)
```

```python
# go through sample.json and ask questions to gpt
import json
sample = json.load(open("sample.json"))
for i in sample:
    answers = []
    mod = []
    print(i["nq_question"])
    for _ in range(3):
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": i['nq_question'],
                }
            ],
            model="gpt-4o-2024-08-06",
        )
        answers.append(chat_completion.choices[0].message.content)

        sys_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "system",
                    "content": system_prompt
                },
                {
                    "role": "user",
                    "content": i['nq_question'],
                }
            ],
            model="gpt-4o-2024-08-06",
        )
        mod.append(sys_completion.choices[0].message.content)

        # print("-"*20)
    print("-"*20)
    
    # show the difference between the answers, characterwise
    for a in range(len(answers)):
        print(answers[a])
        print(mod[a])
        print("-"*20)


    print("="*20)
    

# answer varies, but is constant for questions that are 
```
